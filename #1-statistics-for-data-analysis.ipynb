## Pengumpulan Data
***
Pengumpulan data dari suatu populasi dalam ilmu statistika dibedakan dengan dua cara, sensus dan sampling (Sudjana, 1989). <br>
&#9658; Sensus: pengumpulan data yang dilakukan dengan cara mengambil informasi yang dibutuhkan dari semua anggota populasi. <br>
&#9658; Sampling: pengumpulan data yang hanya berasal dari hanya sebagian anggota populasi (subset population).

Perbedaan ini tentu akan menyebabkan perbedaan dalam metode pengambilan keputusan. Apapun metode pengumpulan datanya, pengambilan keputusan tetap ditujukan untuk menyimpulkan kondisi real dari populasi.

•	Ketika pengumpulan data dilakukan secara *sensus* maka pengambilan keputusan terhadap populasi menjadi sah ketika hanya dilakukan secara deskriptif melalui gambaran data, disebut *statistika deskriptif* <br>
•	Ketika pengumpulan data dilakukan secara *sampling* maka pengambilan keputusan terhadap populasi tidak cukup dengan deskriptif, melainkan harus dilakukan dengan *statistika inferensia*

#### Mean
***
- The sample mean, also called the sample arithmetic mean or simply the average, is the arithmetic average of all the items in a dataset. The mean of a dataset 𝑥 is mathematically expressed as Σᵢ𝑥ᵢ/𝑛, where 𝑖 = 1, 2, …, 𝑛. In other words, it’s the sum of all the elements 𝑥ᵢ divided by the number of items in the dataset 𝑥.

- The mean, or average, is a measure that attempts to summarize data by calculating the central position using the formula:

- The mean can only be applied to a set of quantitative data. Using the student performance dataset, it is only possible to get the average for the student scores.
- The mean is a very simple measure to calculate and it can help provide a value that can summarize the data. However, the mean is very sensitive to outliers/extreme values. For instance, the income of 5 men selected at random might be: <br>
&#9632; Rp. 1 juta <br>
&#9632; Rp 2 juta <br>
&#9632; Rp 2,5 juta <br>
&#9632; Rp 3 juta <br>
&#9632; Rp 150 juta <br>
- The mean that would be obtained from this data would not be a sufficient representation of it. For data with a lot of extreme values, the mean might not be your most preferred choice.

```python
# from scratch
def mean(list1):
    total = math.fsum(list1)
    n = len(list1)
    mean = total/n
    return mean

values = [1, 2, 3, 4, 5]
print('From Scratch:', mean(values))

# Using Numpy
print('Using Numpy:', np.mean(values))

# Using Pandas for the Student Data
data = pd.read_csv("datasets/student_performance_dataset/StudentsPerformance.csv")
avg_reading = data['reading score'].mean()
print('Using Pandas:', avg_reading)
```

#### Weighted Mean
***
- The weighted mean, also called the weighted arithmetic mean or weighted average, is a generalization of the arithmetic mean that enables you to define the relative contribution of each data point to the result.
- You define one weight 𝑤ᵢ for each data point 𝑥ᵢ of the dataset 𝑥, where 𝑖 = 1, 2, …, 𝑛 and 𝑛 is the number of items in 𝑥. Then, you multiply each data point with the corresponding weight, sum all the products, and divide the obtained sum with the sum of weights: Σᵢ(𝑤ᵢ𝑥ᵢ) / Σᵢ𝑤ᵢ.

- The weighted mean is very handy when you need the mean of a dataset containing items that occur with given relative frequencies. For example, say that you have a set in which 20% of all items are equal to 2, 50% of the items are equal to 4, and the remaining 30% of the items are equal to 8. 

```python
0.2 * 2 + 0.5 * 4 + 0.3 * 8
```

Here, you take the frequencies into account with the weights. With this method, you don’t need to know the total number of items.

```python
x = [8.0, 1, 2.5, 4, 28.0]
w = [0.1, 0.2, 0.3, 0.25, 0.15]
wmean = sum(w[i] * x[i] for i in range(len(x))) / sum(w)
wmean
```

However, if you have large datasets, then NumPy is likely to provide a better solution. You can use np.average() to get the weighted mean of NumPy arrays or pandas Series

```python
y, z, w = np.array(x), data['math score'].head().values, np.array(w)
wmean = np.average(z, weights=w)
wmean
```

#### Median
***
- The median is another measure of central tendency that finds a central value for a set of quantitative data. The median is also very easy to compute and it is not affected by outliers like the mean.
- However, the median does not consider all the data points in the provided data. It emphasizes the position rather than the values and is sometimes referred to as a measure of location/position.

- For example, if you have the data points 2, 4, 1, 8, and 9, then the median value is 4, which is in the middle of the sorted dataset (1, 2, 4, 8, 9). If the data points are 2, 4, 1, and 8, then the median is 3, which is the average of the two middle elements of the sorted sequence (2 and 4). The following figure illustrates this:


#### Mean vs Median
***

- The data points are the green dots, and the purple lines show the median for each dataset. The median value for the upper dataset (1, 2.5, 4, 8, and 28) is 4. If you remove the outlier 28 from the lower dataset, then the median becomes the arithmetic average between 2.5 and 4, which is 3.25.



- The main difference between the behavior of the mean and median is related to dataset outliers or extremes. The mean is heavily affected by outliers, but the median only depends on outliers either slightly or not at all. 



- If you increase its value (move it to the right), then the mean will rise, but the median value won’t ever change.
- If you decrease its value (move it to the left), then the mean will drop, but the median will remain the same until the value of the moving point is greater than or equal to 4.

#### Persentile
***
- Percentiles are values that separate the data into 100 equal parts.
- For example, The 95th percentile separates the lowest 95% of the values from the top 5%
- For example, the 75th percentile represents the point below which 75% of the data falls.

1. The 25th percentile (P25%) is the same as the first quartile (Q1).
2. The 50th percentile (P50%) is the same as the second quartile (Q2) and the median.
3. The 75th percentile (P75%) is the same as the third quartile (Q3)

- The formula to calculate a percentile is as follows:
- P = ((n+1) * k/100)th value in the ordered dataset, where n is the number of data points and k is the desired percentile (e.g. 75th percentile)
- If P is not an integer, round up to the next integer and take the value at that index in the ordered dataset.
- Note that the formula assumes the dataset is already ordered, 

```python
# using numpy
# create a sample dataset
data = np.array([3, 5, 7, 8, 9, 11, 13, 15, 16, 20])

# calculate the 75th percentile using numpy's percentile function
p75 = np.percentile(data, 75)

# print the 75th percentile value
print("75th percentile:", p75)
```

```python
# using pandas 
# create a sample dataset
data = pd.Series([3, 5, 7, 8, 9, 11, 13, 15, 16, 20])

# calculate the 75th percentile using pandas' quantile function
p75 = data.quantile(0.75)

# print the 75th percentile value
print("75th percentile:", p75)
```

#### Rentang Antar Kuartil
***
- Rentang antar kuartil (interquatile range) adalah jarak antara perempat data pertama dengan perempat data ketiga
- Rentang antar kuartil merupakan selisih antara Q3 dan Q1. Biasanya disebut dengan notasi IQR.
- Kegunaan dari ukuran ini adalah penggunaannya pada identifikasi data pencilan dengan BoxPlot.
- Untuk menggambarkan sebara data secara jelas, kita tidak bisa menggunakan nilai IQR saja. Kita memerlukan ***five number summary*** yang berisi Minimum, Q1, Median, Q3, dan Maksimum
***
**Boxplot (Intermesso)**<br>
- Boxplot adalah grafik yang berupa garis , kotak, dan terkadang dilengkapi dengan satu atau lebih titik.
- Ujung bawah garis menyatakan Minimum
- Ujung bawah kotak menyatakan Q1
- Garis di dalam kotak menyatakan Median
- Ujung atas kotak menyatakan Q3
- Ujung atas garis menyatakan Maksimum

***
**Outliers Detection with Boxplot** <br>
IQR bisa digunakan untuk deteksi data pencilan pada suatu kumpulan data. <br>
- Suatu data dikatakan sebagai pencilan apabila **lebih kecil** dari **1,5 IQR dibawah Q1** atau **lebih besar** dari **1,5 IQR diatas Q3**




```python
# create a sample dataset
data = np.array([3, 5, 7, 8, 9, 11, 13, 15, 16, 20, 170])

# create a box plot of the data
fig, ax = plt.subplots()
ax.boxplot(data)

# set the title and axis labels
ax.set_title('Box Plot of Sample Data')
ax.set_xlabel('Data')
ax.set_ylabel('Values')

# calculate and display the five-number summary
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

# print the results
print("Minimum:", data.min())
print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
print("Maximum:", data.max())
print("Lower Whisker:", lower_whisker)
print("Upper Whisker:", upper_whisker)

# display the plot and the five-number summary
plt.show()
```

#### Variance
***
- variance menyatakan ukuran seberapa jauh kumpulan bilangan tersebar. Makin kecil nilai variance maka setiap data makin dekat dengan nilai rata-ratanya
- The sample variance quantifies the spread of the data. It shows numerically how far the data points are from the mean. You can express the sample variance of the dataset 𝑥 with 𝑛 elements mathematically as 𝑠² = Σᵢ(𝑥ᵢ − mean(𝑥))² / (𝑛 − 1), where 𝑖 = 1, 2, …, 𝑛 and mean(𝑥) is the sample mean of 𝑥. 
- The following figure shows you why it’s important to consider the variance when describing datasets:

- There are two datasets in this figure:
1. Green dots: This dataset has a smaller variance or a smaller average difference from the mean. It also has a smaller range or a smaller difference between the largest and smallest item.
2. White dots: This dataset has a larger variance or a larger average difference from the mean. It also has a bigger range or a bigger difference between the largest and smallest item.

#### Simpangan Baku
***
- The sample standard deviation is another measure of data spread. It’s connected to the sample variance, as standard deviation, 𝑠, is the **positive square root of the sample variance**. 
- The standard deviation is often more convenient than the variance because it has the same unit as the data points.
***

#### Variance and Simpangan Baku
***
- You calculate the population variance similarly to the sample variance. However, you have to use 𝑛 in the denominator instead of 𝑛 − 1: Σᵢ(𝑥ᵢ − mean(𝑥))² / 𝑛. In this case, 𝑛 is the number of items in the entire population. 
- You can get the population variance similar to the sample variance, with the following differences:
1. Replace (n - 1) with n in the pure Python implementation.
2. Use statistics.pvariance() instead of statistics.variance().
3. Specify the parameter ddof=0 if you use NumPy or pandas. In NumPy, you can omit ddof because its default value is 0.
***
- The population standard deviation refers to the entire population. It’s the positive square root of the population variance. You can calculate it just like the sample standard deviation, with the following differences:
1. Find the square root of the population variance in the pure Python implementation.
2. Use statistics.pstdev() instead of statistics.stdev().
3. Specify the parameter ddof=0 if you use NumPy or pandas. In NumPy, you can omit ddof because its default value is 0.

- As you can see, you can determine the standard deviation in Python, NumPy, and pandas in almost the same way as you determine the variance. You use different but analogous functions and methods with the same arguments.
***

**NOTE!!** that you should always be aware of whether you’re working with a sample or the entire population whenever you’re calculating the variance!

***
**ddof**: delta degree of freedom <br>
**ddof=1** refer to sample variance or sample standard deviation <br>
**ddof=0** refer to population variance or population standard deviation

```python
# using numpy

# create a sample dataset
data = np.array([3, 5, 7, 8, 9, 11, 13, 15, 16, 20])

# calculate variance and standard deviation using numpy
variance = np.var(data)
std_dev = np.std(data)

# print the variance and standard deviation
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

```python
# using pandas
# create a sample dataset
data = pd.Series([3, 5, 7, 8, 9, 11, 13, 15, 16, 20])

# calculate variance and standard deviation using pandas
variance = data.var()
std_dev = data.std()

# print the variance and standard deviation
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

**Analisis** <br>
**Pertanyaan?** Kenapa hasil di atas bisa berbeda padahal menggunakan data yang sama?
***
- In numpy, the variance is calculated using the formula:
variance = mean((x - x.mean())**2)
- In pandas, the variance is calculated using the formula:
variance = sum((x - x.mean())**2) / (n - ddof)

where x is the input series, n is the number of elements in the series, and ddof is the degrees of freedom adjustment (by default, ddof=1). <br>

- Similarly, for standard deviation calculations, numpy uses the formula: std_dev = sqrt(mean((x - x.mean())**2))
- and pandas uses the formula: std_dev = sqrt(sum((x - x.mean())**2) / (n - ddof))

#### Skewness
***
- The sample skewness measures the asymmetry of a data sample. <br>
- There are several mathematical definitions of skewness. A simpler expression is Σᵢ(𝑥ᵢ − mean(𝑥))³ 𝑛 / ((𝑛 − 1)(𝑛 − 2)𝑠³), where 𝑖 = 1, 2, …, 𝑛 and mean(𝑥) is the sample mean of 𝑥. The skewness defined like this is called the **adjusted Fisher-Pearson standardized moment coefficient**.
- A distribution can be symmetric, meaning it has a similar shape on both sides of its center, or it can be skewed to one side or the other. Skewness can be positive, negative, or zero, depending on the direction and degree of the skew.
- If the skewness value is positive, the distribution is skewed to the right (tail on the right side), while if the skewness value is negative, the distribution is skewed to the left (tail on the left side). A skewness value of zero indicates a symmetric distribution.
***



```python
# implement skewness using numpy
# create a sample dataset
data = np.array([3, 5, 7, 8, 9, 11, 13, 15, 16, 20])

# calculate skewness using numpy and scipy
skewness_np = np.mean((data - np.mean(data))**3) / np.std(data)**3

# print the skewness values
print("Skewness (numpy):", skewness_np)
```

```python
# implement using pandas
# create a sample dataset
data = pd.Series([3, 5, 7, 8, 9, 11, 13, 15, 16, 20])

# calculate skewness using pandas
skewness = data.skew()

# print the skewness value
print("Skewness:", skewness)
```

```python
# Generate a sample of 100 data points from a normal distribution
data = np.random.normal(0, 1, 100)

# Calculate the skewness using NumPy
skewness = np.mean((data - np.mean(data))**3) / np.power(np.var(data), 1.5)

# Plot the skewness
fig, ax = plt.subplots()
ax.axvline(skewness, color='r', linestyle='--')
ax.annotate(f"Skewness: {skewness:.2f}", xy=(skewness, 0.5), xytext=(skewness, 0.5),
            fontsize=12, ha='center', va='center', color='r')
ax.hist(data, bins=20, alpha=0.5)
ax.set_xlabel('Data')
ax.set_ylabel('Frequency')
ax.set_title('Skewness Plot')
plt.show()
```

### Pengukuran Korelasi antar pasangan pada data
***
Memeriksa hubungan antara elemen yang sesuai dari dua variabel dalam kumpulan data. Semisal ada dua variabel X dan Y dan di variabel X ada elemen (x1,x2,...,xn) dan di variabel Y ada elemen (y1,y2,...,yn). Maka ada sejumlah n pasang elemen yang bersesuaian (x1,y1), (x2,y2),...,(xn,yn).
Ada tiga jenis pengukuran korelasi:
1. Positive correlation
2. Negative correlation
3. Weak or no correlation



Dua statistik untuk mengukur korelasi dataset adalah **covariance** dan **correlation coefficient**

#### Covariance
***
Sample Covariance adalah ukuran yang mengukur kekuatan dan arah hubungan antara sepasang variabel.
1. Jika korelasi positif, maka covariance positif. Hubungan yang kuat menandakan nilai yang tinggi dari covariance
2. Jika korelasi negatif, maka covariance negatif.
3. Jika korelasi lemah (close to zero), maka covariance mendekati 0
***


```python
# inisialisasi data 
x = list(range(-10, 11))
y = [0, 2, 2, 2, 2, 3, 3, 6, 7, 4, 7, 6, 6, 9, 4, 5, 5, 10, 11, 12, 14]
x_, y_ = np.array(x), np.array(y)
x__, y__ = pd.Series(x_), pd.Series(y_)
```

```python
# hitung covariance matrix dengan NumPy
cov_matrix = np.cov(x_,y_)
cov_matrix
```

Elemen cov_matrix(0,0) menandakan covariance dari x dengan x, begitupun cov_matrix(1,1) menandakan covariance dari y dengan y. Elemen cov_matrix(0,1) dan cov_matrix(1,0) adalah covariance aktual antara x dan y

```python
# hitung variance dari masing-masing variabel dengan numpy
print("Variance variabel x=", x_.var(ddof=1))
print("Variance variabel y=", y_.var(ddof=1))
```

As you can see, the variances of x and y are equal to cov_matrix[0, 0] and cov_matrix[1, 1], respectively.

```python
# calculate covariance using pandas using .cov() method
cov_xy = x__.cov(y__)
cov_yx = y__.cov(x__)
print("Covariance xy = ", cov_xy)
print("Covariance yx = ", cov_yx)
```

#### Correlation Coefficient
***
**Definition** <br>
- Correlation coefficient is a statistical measure that quantifies the degree of association between two variables. It indicates the strength and direction of a linear relationship between two variables, where the values range between -1 and +1. <br>
- A positive correlation coefficient means that the two variables move together in the same direction, while a negative correlation coefficient means that the two variables move in opposite directions.

**Formula** <br>
- The correlation coefficient is often denoted by the symbol "*r*". It is calculated by dividing the covariance of the two variables by the product of their standard deviations:

|r = cov(X, Y) / (std(X) * std(Y))|
|---|

where cov(X,Y) is the covariance between X and Y, and std(X) and std(Y) are the standard deviations of X and Y, respectively.

**Correlation Indication** <br>
1. The value 𝑟 > 0 indicates positive correlation.
2. The value 𝑟 < 0 indicates negative correlation.
3. The value r = 1 is the maximum possible value of 𝑟. It corresponds to a perfect positive linear relationship between variables.
4. The value r = −1 is the minimum possible value of 𝑟. It corresponds to a perfect negative linear relationship between variables.
5. The value r ≈ 0, or when 𝑟 is around zero, means that the correlation between variables is weak.

**Purpose** <br>
The correlation coefficient is widely used in statistical analysis, data science, and machine learning to investigate the relationships between variables and to make predictions. It can be used to identify patterns, to test hypotheses, and to build predictive models.

```python
# using pure Python
n = len(x)
mean_x, mean_y = sum(x) / n, sum(y) / n
var_x = sum((item - mean_x)**2 for item in x) / (n - 1)
var_y = sum((item - mean_y)**2 for item in y) / (n - 1)
std_x, std_y = var_x ** 0.5, var_y ** 0.5
r = cov_xy / (std_x * std_y)
r
```

```python
# using numpy
corr_matrix = np.corrcoef(x_, y_)
corr_matrix
```

```python
# using pandas Series
r = x__.corr(y__)
print('Correlation Coefficient xy = ', r)
r = y__.corr(x__)
print('Correlation Coefficient yx = ', r)
```

You should call .corr() on one Series object and pass the other object as the first argument.

#### Axes in Multidimensional Array

```python
a = np.array([[1, 1, 1],
              [2, 3, 1],
              [4, 9, 2],
              [8, 27, 4],
              [16, 1, 1]])
a
```

```python
# apply some statistics function into it
print(np.mean(a))
print(a.mean())
print(np.median(a))
print(a.var(ddof=1))
```

As you can see, you get statistics (like the mean, median, or variance) across all data in the array a. <br>
Sometimes, this behavior is what you want, but in some cases, you’ll want these quantities calculated for each row or column of your 2D array.

***
The functions and methods you’ve used so far have one optional parameter called axis, which is essential for handling 2D data. axis can take on any of the following values: <br>
1. **axis=None** says to calculate the statistics across all data in the array. The examples above work like this. This behavior is often the default in NumPy.
2. **axis=0** says to calculate the statistics across all rows, that is, for each column of the array. This behavior is often the default for SciPy statistical functions.
3. **axis=1** says to calculate the statistics across all columns, that is, for each row of the array.

```python
print(a.mean(axis=0))
print(a.mean(axis=1))
print(a.var(axis=1, ddof=1))
```

In practice, the names of the columns matter and should be descriptive. The names of the rows are sometimes specified automatically as 0, 1, and so on. You can specify them explicitly with the parameter index, though you’re free to omit index if you like.

```python
print(df.mean())
print("="*30)
print(df.var())
```
